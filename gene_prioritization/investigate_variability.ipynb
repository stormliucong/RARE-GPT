```python
import pandas as pd
```

```python
gpt_evaluation_df = pd.read_csv('/phi_home/cl3720/phi/RESCUE/RARE-GPT/gene_prioritization/Experiment_003subset_eval_table.csv')
```

```python
# calculate how many times the completeness are all equal among different iterations per group
completeness_variability_df = gpt_evaluation_df.groupby(['sample_id','gpt_version','prompt','top_n']).apply(lambda x: x['completeness'].nunique()).reset_index()
print((completeness_variability_df[0] == 1).sum())
print((completeness_variability_df[0] == 2).sum())
completed_df = gpt_evaluation_df[gpt_evaluation_df['completeness'] == 1]
completed_df = completed_df[['sample_id','gpt_version','prompt','top_n']]
completed_completeness_variability_df = completed_df.merge(gpt_evaluation_df).groupby(['sample_id','gpt_version','prompt','top_n']).apply(lambda x: x['completeness'].nunique()).reset_index()
print((completed_completeness_variability_df[0] == 1).sum())
print((completed_completeness_variability_df[0] == 2).sum())
```

```python
# calculate how many times the completeness are all equal among different iterations per group
accuracy_variability_df = gpt_evaluation_df[gpt_evaluation_df['accuracy'].notna()].groupby(['sample_id','gpt_version','prompt','top_n']).apply(lambda x: x['accuracy'].nunique()).reset_index()
print((accuracy_variability_df[0] == 1).sum())
print((accuracy_variability_df[0] == 2).sum())
accurate_df = gpt_evaluation_df[gpt_evaluation_df['accuracy'] == 1]
accurate_df = accurate_df[['sample_id','gpt_version','prompt','top_n']]
accurate_accuracy_variability_df = accurate_df.merge(gpt_evaluation_df[gpt_evaluation_df['accuracy'].notna()]).groupby(['sample_id','gpt_version','prompt','top_n']).apply(lambda x: x['accuracy'].nunique()).reset_index()
print((accurate_accuracy_variability_df[0] == 1).sum())
print((accurate_accuracy_variability_df[0] == 2).sum())
```
